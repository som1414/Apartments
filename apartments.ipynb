- –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏;
- –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏;
- –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏;
- –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–æ–º–Ω–∞—Ç;
- —ç—Ç–∞–∂–∞, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞ (–ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π);
- –¥–∞—Ç—ã —Ä–∞–∑–º–µ—â–µ–Ω–∏—è (–¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü, –≥–æ–¥). 

**–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏**

```python
def cor(column):
    print('–ö–æ—Ä–µ–ª–ª—è—Ü–∏—è —Ü–µ–Ω—ã –æ—Ç', column, df['last_price'].corr(df[column]))
    df.plot(x=column, y='last_price', kind='scatter', grid=True, figsize=(15,5), alpha=0.3)
    plt.show()

cor('total_area')
```

**–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏**

#### 4.4 –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞.


–ü–æ—Å—á–∏—Ç–µ–º —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π. –í—ã–¥–µ–ª–∏–º –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞.

```python
top10 = df.pivot_table(index = 'locality_name', values = 'price_metr', aggfunc=['count', 'mean'])
top10.columns = ['count', 'price_metr']
top10 = top10.sort_values(by='count', ascending=False).head(10)

display(top10)

print('–ù–∞—Å–µ–ª—ë–Ω–Ω—ã–π –ø—É–Ω–∫—Ç —Å c–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞:')
print(top10.loc[top10['price_metr'] == top10['price_metr'].max()])
print()
print('–ù–∞—Å–µ–ª—ë–Ω–Ω—ã–π –ø—É–Ω–∫—Ç c c–∞–º–æ–π –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞:')
print(top10.loc[top10['price_metr'] == top10['price_metr'].min()])
```
